```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
ax1.bar(app_train.CODE_GENDER.unique(),app_train.CODE_GENDER.value_counts())
ax1.set_title('CODE_GENDER')
ax2.bar(app_train.FLAG_OWN_CAR.unique(),app_train.FLAG_OWN_CAR.value_counts())
ax2.set_title('FLAG_OWN_CAR')
ax3.bar(app_train.FLAG_OWN_REALTY.unique(),app_train.FLAG_OWN_REALTY.value_counts())
ax3.set_title('FLAG_OWN_REALTY')
plt.show()
```

```python
ct = pd.crosstab(app_train.CODE_GENDER, app_train.FLAG_OWN_CAR, margins = True, dropna=True)
print(ct)
```

```python
data = app_train[['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE']]
plt.scatter(data.AMT_INCOME_TOTAL, data.AMT_CREDIT)
plt.xlabel('INCOME', fontsize=13)
plt.ylabel('CREDIT', fontsize=13)
plt.title('INCOME vs. CREDIT', fontsize=17)
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(data, alpha=0.2, figsize=(6, 6), diagonal='kde') 
```
