```python
# import packges
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv
train = pd.read_csv('train.csv', header=None)
trainLabel = pd.read_csv('trainLabels.csv', header=None)
test = pd.read_csv('test.csv', header=None)
```

### Logistic Regression

```python
# preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X, y = train, np.ravel(trainLabel)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver = 'newton-cg')
classifier.fit(X_train, y_train)
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
# Making the Confusion Matrix, accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
acc
```

```python
cm
```

#### Adjust the hyper coefficient

```python
from sklearn.model_selection import train_test_split, KFold, GridSearchCV

# Combination
max_iter = [1, 5, 10, 15, 20, 25]
param_grid = dict(max_iter = max_iter)

# Grid search
grid_search = GridSearchCV(classifier, param_grid, scoring="accuracy", n_jobs=-1, verbose=1)
grid_result = grid_search.fit(X_train, y_train)

# 印出最佳結果與最佳參數
print("Best Accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
