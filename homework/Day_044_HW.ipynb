### 1.

```python
from sklearn import datasets, metrics
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 讀取鳶尾花資料集
iris = datasets.load_iris()

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25, random_state=4)
```

```python
# 建立模型 (max_depth = 4)
x = []
y = []
for i in range(1,51):
    clf = RandomForestClassifier(n_estimators=i, max_depth=4)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    x.append(i)
    y.append(metrics.accuracy_score(y_test, y_pred))
plt.plot(x, y)
plt.show()
y_max = np.where(y==max(y))
print(y_max +  np.ones(len(y_max)).reshape(1,-1))
```

```python
ma = metrics.confusion_matrix(y_test, y_pred)
ma
```

### 2. (Boston)

```python
# 讀取boston資料集
boston = datasets.load_boston()

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.25, random_state=4)
```

```python
# 建立模型 (max_depth = 4)
x = []
y = []
for i in range(1,51):
    clf = RandomForestRegressor(n_estimators=i, max_depth=4)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    x.append(i)
    y.append(metrics.mean_absolute_error(y_test, y_pred))
plt.plot(x, y)
plt.show()
y_min = np.where(y==min(y))
print(y_min +  np.ones(len(y_min)).reshape(1,-1))
print(min(y))
```

```python
# 建立模型 (max_depth = 8)
x = []
y = []
for i in range(1,51):
    clf = RandomForestRegressor(n_estimators=i, max_depth=8)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    x.append(i)
    y.append(metrics.mean_absolute_error(y_test, y_pred))
plt.plot(x, y)
plt.show()
y_min = np.where(y==min(y))
print(y_min +  np.ones(len(y_min)).reshape(1,-1))
print(min(y))
```
