### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets
from sklearn import decomposition
from sklearn import datasets

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(5)
%matplotlib inline

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

```python
import numpy as np
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
pca = decomposition.PCA(n_components=3)
```

```python
# decomposition.pca:　https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
pca.fit(X)
X = pca.transform(X)

# plt.cl: https://stackoverflow.com/questions/16661790/difference-between-plt-close-and-plt-clf
fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.cla()

# text3D: https://matplotlib.org/3.1.1/api/_as_gen/mpl_toolkits.mplot3d.axes3d.Axes3D.html?highlight=ax%20text3d#mpl_toolkits.mplot3d.axes3d.Axes3D.text3D

for name, label in [('Group_0', 0), ('Group_1', 1), ('Group_2', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean()+1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
# np.choose:　https://docs.scipy.org/doc/numpy/reference/generated/numpy.choose.html
# y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral,
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
