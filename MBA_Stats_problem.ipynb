```python
# Logistic Regression
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing training dataset
df_train = pd.read_excel('Train.xlsx', sheetname='Survey Data')
df_train = df_train.set_index('ID')
df_train.head()
```

```python
# Importing testing dataset
df_test = pd.read_excel('Test.xlsx', sheetname='Sheet1')
df_test = df_test.set_index('ID')
df_test.head()
```

```python
# No Scaling-Buy1
X_train_1 = df_train.iloc[:, 1:11].values
y_train_1 = df_train.iloc[:, 12].values
X_test_1 = df_test.iloc[:, :10].values

# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier_1 = LogisticRegression()
classifier_1.fit(X_train_1, y_train_1)

# Predict
y_pred_1 = classifier_1.predict(X_test_1)
y_pred_1_prob = classifier_1.predict_proba(X_test_1)
```

```python
y_pred_1_prob[:, 1]
```

```python
# No Scaling-Buy2
X_train_2 = df_train.iloc[:, 1:11].values
y_train_2 = df_train.iloc[:, 13].values
X_test_2 = df_test.iloc[:, :10].values

# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier_2 = LogisticRegression()
classifier_2.fit(X_train_2, y_train_2)

# Predict
y_pred_2 = classifier_2.predict(X_test_2)
y_pred_2_prob = classifier_2.predict_proba(X_test_2)
```

```python
y_pred_2_prob[:, 1]
```

```python
# Scaling-Buy1
X_train_1 = df_train.iloc[:, 1:11].values
y_train_1 = df_train.iloc[:, 12].values
X_test_1 = df_test.iloc[:, :10].values

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_1_1 = sc.fit_transform(X_train_1)
X_test_1_1 = sc.fit_transform(X_test_1)

# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier_1_1 = LogisticRegression()
classifier_1_1.fit(X_train_1_1, y_train_1)

# Predict
y_pred_1_1 = classifier_1_1.predict(X_test_1_1)
y_pred_1_1_prob = classifier_1_1.predict_proba(X_test_1_1)
```

```python
y_pred_1_1_prob[:, 1]
```

```python
np.where(y_pred_1_1 == 1)
```

```python
# Scaling-Buy2
X_train_2 = df_train.iloc[:, 1:11].values
y_train_2 = df_train.iloc[:, 13].values
X_test_2 = df_test.iloc[:, :10].values

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_2_1 = sc.fit_transform(X_train_2)
X_test_2_1 = sc.fit_transform(X_test_2) 

# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier_2_1 = LogisticRegression()
classifier_2_1.fit(X_train_2_1, y_train_2)

# Predict
y_pred_2_1 = classifier_2_1.predict(X_test_2_1)
y_pred_2_1_prob = classifier_2_1.predict_proba(X_test_2_1)
```

```python
y_pred_2_1_prob[:, 1]
```

```python
np.where(y_pred_2_1 == 1)
```

```python
# Report Dataframe
data = {'Buy_1':y_pred_1_1_prob[:, 1],
        'Buy_2':y_pred_2_1_prob[:, 1]}
df = pd.DataFrame(data)
df['ID'] = pd.read_excel('Train.xlsx', sheetname='Survey Data')['ID']
df['Score'] = 0.5* df.Buy_1 + 0.5* df.Buy_2 # 假設兩個PROGRAM權重一樣
df = df.set_index('ID')
df.head()
```

```python
# Report Score sorting
Report = df.sort_values('Score',ascending=False)
Report.head()
```

```python
# Report Export
Report = Report[:55]
Report.to_excel("Report.xlsx",sheet_name='Answer')
```
